### Variables

Variable is open of abstractions in the math and programming world. In Python, it means something like a container for storing data values. Each variable can store only one data value at the time! (but it could be one set of data values, which will be discussed later).

* we can assign values to variables using an assignment operator (`=`)
* we assign variable value to a variable name in a following expression: variable_name = variable_value
* some rules:
  * upper- and lower-case characters, digits and underscores(_) are allowed
  * variable names cannot start with a digit
* please be semantic
* [snake_case](https://en.wikipedia.org/wiki/Snake_case) is a convention for variable naming in python

````python
hello_world = 'Hello world!' # we assign literal value 'Hello World!' to a variable named 'hello_world'
print(hello_world)
i = 5
i = i+1 # we assign new value to i
print(i)
i += 1 # we assign i to i+i
print(i)
i -= 1
print(i)
i *= 2 # we assign i to i * 2
print(i)
````

Let's rewrite the last if-else structure to convert 10-grade HSE estimation scale to 5-grade one.

````python
your_grade = int(input("Enter your 10-scale grade here"))
````

````python
# version 1 - quick convertor
if your_grade >= 8:
    print("Your score is 5!")
elif your_grade >= 6:
    print("Your score is 4!")
elif your_grade >= 4:
    print("Your score is 3!")
else:
    print("You failed!")
````

But what if we enter 11?

````python
# version 1.1 - taking into account all possible inputs
if your_grade > 10:
    print("This is not possible! Check you grade once again.")
elif your_grade >= 8:
    print("Your score is 5!")
elif your_grade >= 6:
    print("Your score is 4!")
elif your_grade >= 4:
    print("Your score is 3!")
elif your_grade >= 0:
    print("You failed!")
else:
    print("Negatives are impossible! Check you grade once again.")
````

````python
# version 1.2 - learning to optimize code with ifelse inside ifelse
if your_grade > 10 or your_grade < 0:
    print("This is not possible! Check you grade once again.")
else:
    if your_grade >= 8:
        print("Your score is 5!")
    elif your_grade >= 6:
        print("Your score is 4!")
    elif your_grade >= 4:
        print("Your score is 3!")
    else:
        print("You failed!")
````

As you can see, all these versions are correct ways to convert 10-scale grade to 5-scale one. However, sometimes it's important to take into account all possible user inputs to avoid inaccuracies and errors. Make sure you considered different possible inputs so the logic of your code won't crash.

````python
# ex. write the flow to guess user's number from 1 to 7
# version 1: hardcode

num = int(input("Guess the single digit integer!"))

if num < 1 or num > 7:
    print('Illegal number!')
    
else:
    guess = 7//2+1
    print("Your number is ", str(guess), "?")

    answer = input('enter "yes", "more", or "less"')
    if answer == "yes":
        print("Hooray!")
        
    elif answer == "more":
        guess += guess//2
        print("Your number is ", str(guess), "?")
        answer = input('enter "yes", "more", or "less"')
        if answer == "yes":
            print("Hooray!")
        elif answer == "more":
            print("Your number is 7!")
        else:
            print("Your number is 5!")

    else:
        guess -= guess//2
        print("Your number is ", str(guess), "?")
        answer = input('enter "yes", "more", or "less"')
        if answer == "yes":
            print("Hooray!")
        elif answer == "more":
            print("Your number is 3!")
        else:
            print("Your number is 1!")
````
