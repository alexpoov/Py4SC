#### Speaking in RegEx

You may ask: *"why do we need this module, if we have the same functionality in string methods?"*. Well, only this module allows to work with [Regular Expressions](https://en.wikipedia.org/wiki/Regular_expression) language. It looks something like this:

```
/^[a-zA-Z0-9.!#$%&*+/=?^_`{|}~-]+@[a-zA-Z0-9-]+(?:\.[a-zA-Z0-9-]+)*$/
```

For clarity, in this topic we will use the [regex101](https://regex101.com/) site.

````python
aland  = "In 1920, Finland granted wide-reaching cultural and political autonomy to the Aland Islands. The League of Nations considered these measures as satisfying demands to protect the Swedish language and culture there."
print(aland)
````

````python
re.findall('.', aland)  # dot means 'symbol' on regex
re.findall('\.', aland)  # dot with backslash will look for dot.
````

````python
re.findall('[abc]', aland)  # to search for all a and b and c letters
re.findall('[o][n][a-z][a-z]', aland)  # all combinations of ...on**... matches
````

````python
re.findall('\d', aland)  # to find any digit
re.findall('\d{4}', aland)  # to capture the year
````

````python
re.findall('[A-Z]', aland)  # to find any capital letter
re.findall('[A-Z][a-z]*', aland)  # to extract words beginning with capital letter
# * means any number of previous pattern matches
````

````python
info = 'Despite eyewitness reports in Dec 1969, it was first seen in Mar 1970'
re.findall('(\d+)', info)  # + means one or more of any digits
re.findall('(\w+) (\d+)', info)  # to extract both month and year (use parenthesis)
````

`ex.` Guess whithout running, which result would be in `(\w+ (\d+))` query?

````python
re.findall('\w*w\w+', 'meow')
````

````python
re.findall('\w*w\w*', 'meow')
````

````python
re.findall('[A-Za-z]*al', aland)  # to capture words ending up with 'al'
re.findall('[\w]*al', aland)  # the same, but shorter 
````
